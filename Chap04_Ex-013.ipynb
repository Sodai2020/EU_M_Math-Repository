```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

#可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

%precision 3
```

```python
import requests, zipfile
from io import StringIO
import io
```

```python
#データがあるURL
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

#データをURLから取得
r = requests.get(url, stream=True)

#zipfileから読み込み展開する
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

```python
student_data_math = pd.read_csv('student-mat.csv', sep=';')
student_data_por = pd.read_csv('student-por.csv',sep=';')
student_data_merge = pd.merge(student_data_math
                                  , student_data_por
                                  , on=['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu'
                                            , 'Fedu', 'Mjob', 'Fjob', 'reason', 'nursery', 'internet']
                                  , how='inner'
                                  , suffixes=('_math', '_por'))

from scipy import stats

print('G2_math_ave : ',student_data_merge.G2_math.mean())
print('G2_por_ave :', student_data_merge.G2_por.mean())

t, p = stats.ttest_rel(student_data_merge.G2_math, student_data_merge.G2_por)
print('p_measure =', p)
```

```python
print('G3_math_ave : ',student_data_merge.G3_math.mean())
print('G3_por_ave : ',student_data_merge.G3_por.mean())

t, p = stats.ttest_rel(student_data_merge.G3_math, student_data_merge.G3_por)
print('p_measure = ',p)
```

# 差があると言えるか? <br>

G2, G3共に, p値が0.01未満なので, 2σで検定するとして, <br>
有意差が0.05の中に入っているので, 双方とも, <br>
差があるという仮説が有意であると言える.<br>
